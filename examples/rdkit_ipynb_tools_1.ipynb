```python
%load_ext autoreload
%autoreload 2

from rdkit.Chem import AllChem as Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
Draw.DrawingOptions.atomLabelFontFace = "DejaVu Sans"
Draw.DrawingOptions.atomLabelFontSize = 18

from rdkit_ipynb_tools import tools, html_templates as html
```

### Molecule Table From a List of Molecules: *show_table()*

```python
mol_list = [mol for mol in Chem.SupplierFromFilename("sdf/testset.sdf")]
```

#### Molecule properties are shown as columns

```python
tools.show_table(mol_list[:3])
```

Put the folder "html" contained in the project rdkit_ipynb_tools into the folder where your IPython notebook resides to have some basic CSS formatting of the table when you write it to file. By default, the file is written to html/mol_table/index.html:

```python
html.write(html.page(tools.mol_table(mol_list)))
```

#### Select rows by clicking on them

When the name of an ID property is passed to the show_table function, the table becomes interactive and rows can be selected by clicking on the structure.  
The value of the respective ID property is inserted into a text box at the bottom and can be used for copy-and-pasting.  
When *Query for Remark* is checked, the user is prompted to enter a remark for each selection (e.g. the reason for selecting this compound).  
ID and remark are tab-separated for easy pasting into spreadsheets.

```python
tools.show_table(mol_list[:5], id_prop="molid")
```

#### Passing a highlight dictionary allows cell highlighting by value

```python
hl = {"logp": "> 3", "molwt": "< 300"}
tools.show_table(mol_list[:5], highlight=hl)
```
